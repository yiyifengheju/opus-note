## 壹丨常见用法

### 1. 迭代器

```python
import tqdm
import time
import random
```

```python
for i in tqdm.tqdm(range(10)):
    time.sleep(0.1)
```

### 2. 手动修改输出格式

```python
out = tqdm.tqdm(['a', 'b', 'c', 'd'])
for item in out:
    time.sleep(0.5)
    out.set_description(f'processing {item}')
```

```python
with tqdm.trange(10) as t:
    for i in t:
        # 设置进度条左边显示的信息
        t.set_description(f"GEN {i}")
        # 设置进度条右边显示的信息
        t.set_postfix(loss=random.random(), gen=random.randint(1, 99), str="h", lst=[1, 2])
        time.sleep(0.1)
```

```python
with tqdm.tqdm(total=100, bar_format="{postfix[0]}{postfix[1][value]:>9.3g}", postfix=["Batch", dict(value=0)]) as t:
    for i in range(100):
        time.sleep(0.05)
        t.postfix[1]["value"] = i / 2
        t.update()
```

### 3. 指定更新size

```python
with tqdm.tqdm(total=10) as pbar:
    for i in range(10):
        time.sleep(0.1)
        pbar.update(1)
```

```python
pbar = tqdm.tqdm(total=10)
for i in range(10):
    time.sleep(0.1)
    pbar.update(1)
pbar.close()
```

### 4. 使用trange替代range

```python
for i in tqdm.trange(10):
    time.sleep(0.1)
```
