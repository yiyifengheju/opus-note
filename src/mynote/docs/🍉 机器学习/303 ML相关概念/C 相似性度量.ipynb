```python
import numpy as np
from scipy.spatial import distance
```

```python
np.random.seed(22)

x = np.random.random(10) * 10
y = np.random.random(10) * 10
print(f'X = {x}\nY = {y}')
```

### 1. æ¬§æ°è·ç¦»ï¼ˆEuclidean Distanceï¼‰

æ¬§æ°è·ç¦»ï¼ˆL2èŒƒæ•°ï¼‰æ˜¯è®¡ç®—ä¸¤ä¸ªå‘é‡ä¹‹é—´çš„ç›´çº¿è·ç¦»ï¼Œå³ä¸¤ä¸ªå‘é‡ä¹‹é—´å¯¹åº”å…ƒç´ å·®çš„å¹³æ–¹å’Œçš„å¹³æ–¹æ ¹ã€‚å®ƒåœ¨å¤„ç†è¿ç»­æ•°æ®æ—¶å¸¸è¢«ä½¿ç”¨ã€‚

$$dist = \sqrt{\sum_{i=1}^n(x_i-y_i)^2}$$

```python
def euclidean_dist(x, y):
    return np.sqrt(np.sum(np.square(x - y)))


dist = euclidean_dist(x, y)
print(f'æ¬§æ°è·ç¦» = {dist}')
dist = np.linalg.norm(x - y)
print(f'L2èŒƒæ•° = {dist}')
dist = distance.euclidean(x, y)
print(f'Scipyæ¬§å¼è·ç¦» = {dist}')
```

### 2. æ›¼å“ˆé¡¿è·ç¦»ï¼ˆManhattan Distanceï¼‰

æ›¼å“ˆé¡¿è·ç¦»ï¼ˆä¹Ÿç§°åŸå¸‚è¡—åŒºè·ç¦»ã€L1èŒƒæ•°ï¼‰æ˜¯è®¡ç®—ä¸¤ä¸ªå‘é‡ä¹‹é—´çš„åŸå¸‚è¡—åŒºè·ç¦»ï¼Œå³ä¸¤ä¸ªå‘é‡ä¹‹é—´å¯¹åº”å…ƒç´ å·®çš„ç»å¯¹å€¼ä¹‹å’Œã€‚å®ƒåœ¨å¤„ç†ç¦»æ•£æ•°æ®æˆ–ç¨€ç–æ•°æ®æ—¶å¸¸è¢«ä½¿ç”¨ã€‚

$$dist = \sum_{i=1}^n|x_i-y_i|$$

```python
def manhattan_dist(x, y):
    return sum(np.abs(x - y))


dist = manhattan_dist(x, y)
print(f'æ›¼å“ˆé¡¿è·ç¦» = {dist}')
dist = np.linalg.norm(x - y, ord=1)
print(f'L1èŒƒæ•° = {dist}')
dist = distance.cityblock(x, y)
print(f'Scipyæ›¼å“ˆé¡¿è·ç¦» = {dist}')
```

### 3. é—µå¯å¤«æ–¯åŸºè·ç¦»ï¼ˆMinkowski Distanceï¼‰

é—µå¯å¤«æ–¯åŸºè·ç¦»æ˜¯æ¬§æ°è·ç¦»å’Œæ›¼å“ˆé¡¿è·ç¦»çš„ä¸€ç§æ¨å¹¿ï¼Œå¯ä»¥é€šè¿‡è°ƒæ•´å‚æ•°æ¥æ§åˆ¶è·ç¦»çš„è®¡ç®—æ–¹å¼ã€‚
$$
d(X, Y) = (|x_1-y_1|^p + |x_2-y_2|^p + ... + |x_n-y_n|^p)^{1/p}
$$
å…¶ä¸­ï¼Œ$p$æ˜¯æ§åˆ¶è·ç¦»è®¡ç®—æ–¹å¼çš„å‚æ•°ï¼Œå½“$p=2$æ—¶ï¼Œé€€åŒ–ä¸ºæ¬§æ°è·ç¦»ï¼›å½“$p=1$æ—¶ï¼Œé€€åŒ–ä¸ºæ›¼å“ˆé¡¿è·ç¦»

```python
def minkowski_dist(x, y, p=1):
    return np.linalg.norm(x - y, ord=p)


p = 3
dist = minkowski_dist(x, y, p=p)
print(f'é—µå¯å¤«æ–¯åŸºè·ç¦» = {dist}')
dist = distance.minkowski(x, y, p=p)
print(f'Scipyé—µå¯å¤«æ–¯åŸºè·ç¦» = {dist}')
```

### 4. ä½™å¼¦ç›¸ä¼¼åº¦ï¼ˆCosine Similarityï¼‰

ä½™å¼¦ç›¸ä¼¼åº¦è¡¡é‡çš„æ˜¯ä¸¤ä¸ªå‘é‡ä¹‹é—´çš„å¤¹è§’ï¼Œè€Œä¸æ˜¯ç›´çº¿è·ç¦»ã€‚å®ƒåœ¨å¤„ç†æ–‡æœ¬æ•°æ®ã€ç¨€ç–æ•°æ®æˆ–é«˜ç»´æ•°æ®æ—¶å¸¸è¢«ä½¿ç”¨ã€‚

$$dist = 1-\frac{x \cdot y}
            {||x||_2 ||y||_2} 
$$
$$ similarity= \frac{\Sigma_ix_iy_i}
            {\sqrt{\Sigma_ix_i^2}\sqrt{\Sigma_iy_i^2}} 
$$

å…¶ä¸­ï¼Œ$XÂ·Y$è¡¨ç¤ºç‚¹ç§¯ï¼ˆå†…ç§¯ï¼‰ï¼Œ$||X||$å’Œ$||Y||$è¡¨ç¤ºL2èŒƒæ•°ï¼ˆé•¿åº¦ï¼‰

ä½™å¼¦ç›¸ä¼¼åº¦çš„å–å€¼èŒƒå›´åœ¨ -1 åˆ° 1 ä¹‹é—´ï¼Œå€¼è¶Šæ¥è¿‘ 1 è¡¨ç¤ºä¸¤ä¸ªå‘é‡è¶Šç›¸ä¼¼ï¼Œå€¼è¶Šæ¥è¿‘ -1 è¡¨ç¤ºä¸¤ä¸ªå‘é‡è¶Šä¸ç›¸ä¼¼ï¼Œå€¼ä¸º 0 è¡¨ç¤ºä¸¤ä¸ªå‘é‡æ­£äº¤ï¼ˆæ— ç›¸ä¼¼æ€§ï¼‰

```python
def cosine_similarity(x, y):
    x = np.asarray(x)
    y = np.asarray(y)
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))


sim = cosine_similarity(x, y)
print(f'ä½™å¼¦ç›¸ä¼¼è·ç¦» = {1 - sim}, ä½™å¼¦ç›¸ä¼¼åº¦ = {sim}')
dist = distance.cosine(x, y)
print(f'Scipyä½™å¼¦ç›¸ä¼¼è·ç¦» = {dist}, ä½™å¼¦ç›¸ä¼¼åº¦ = {1 - dist}')
```

### 5. çš®å°”é€Šç›¸å…³ç³»æ•°ï¼ˆPearson Correlation Coefficientï¼‰

çš®å°”é€Šç›¸å…³ç³»æ•°è¡¡é‡çš„æ˜¯ä¸¤ä¸ªå‘é‡ä¹‹é—´çš„çº¿æ€§ç›¸å…³æ€§ï¼Œå–å€¼èŒƒå›´åœ¨-1åˆ°1ä¹‹é—´ã€‚å®ƒåœ¨å¤„ç†çº¿æ€§ç›¸å…³æ€§é—®é¢˜æ—¶å¸¸è¢«ä½¿ç”¨ã€‚

$$dist =  1 - \frac{(x - \bar{x}) \cdot (y - \bar{y})}
           {{||(x - \bar{x})||}_2 {||(y - \bar{y})||}_2}
$$

$$coor = \frac{(x - \bar{x}) \cdot (y - \bar{y})}
           {{||(x - \bar{x})||}_2 {||(y - \bar{y})||}_2}
$$


```python
def pearson_r(x, y):
    return np.corrcoef(x, y)[0, 1]


def pearson_dist(x, y):
    return 1 - np.corrcoef(x, y)[0, 1]


dist = pearson_dist(x, y)
r = pearson_r(x, y)
print(f'çš®å°”é€Šè·ç¦» = {dist}, çš®å°”é€Šç›¸å…³æ€§ = {r}')
dist = distance.correlation(x, y)
print(f'Scipyçš®å°”é€Šè·ç¦» = {dist}, çš®å°”é€Šç›¸å…³æ€§ = {1 - dist}')
```

### 6. æ°å¡å¾·ç›¸ä¼¼åº¦ï¼ˆJaccard Similarityï¼‰

æ°å¡å¾·ç›¸ä¼¼ç³»æ•°(Jaccard similarity coefficient)ï¼Œä¹Ÿç§°æ°å¡å¾·æŒ‡æ•°(Jaccard Index)ï¼Œç”¨äºè®¡ç®—ä¸¤ä¸ªé›†åˆä¹‹é—´çš„ç›¸ä¼¼æ€§ï¼Œå®ƒè¡¡é‡çš„æ˜¯ä¸¤ä¸ªé›†åˆäº¤é›†çš„å¤§å°ä¸å¹¶é›†çš„å¤§å°ä¹‹é—´çš„æ¯”ä¾‹ã€‚

æ°å¡å¾·è·ç¦»(Jaccard Distance) æ˜¯ç”¨æ¥è¡¡é‡ä¸¤ä¸ªé›†åˆå·®å¼‚æ€§çš„ä¸€ç§æŒ‡æ ‡ï¼Œå®ƒæ˜¯æ°å¡å¾·ç›¸ä¼¼ç³»æ•°çš„è¡¥é›†ï¼Œè¢«å®šä¹‰ä¸º1å‡å»Jaccardç›¸ä¼¼ç³»æ•°ã€‚

$$dist = \frac{c_{TF} + c_{FT}}
        {c_{TT} + c_{FT} + c_{TF}}
$$



```python
dist = distance.jaccard(x, y)
print(f'Scipyæ°å¡å¾·è·ç¦» = {dist}, æ°å¡å¾·ç›¸å…³æ€§ = {1 - dist}')
```

### 7. é©¬æ°è·ç¦»ï¼ˆMahalanobis Distanceï¼‰

é©¬æ°è·ç¦»æ˜¯ä¸€ç§ç”¨äºè¡¡é‡ä¸¤ä¸ªæ ·æœ¬ç‚¹ä¹‹é—´çš„è·ç¦»çš„åº¦é‡æ–¹æ³•ã€‚ä¸æ¬§æ°è·ç¦»ä¸åŒï¼Œé©¬æ°è·ç¦»è€ƒè™‘äº†æ•°æ®çš„åæ–¹å·®ç»“æ„ï¼Œå¯ä»¥æœ‰æ•ˆåœ°å¤„ç†ç‰¹å¾ä¹‹é—´çš„ç›¸å…³æ€§ã€‚

$$dist = \sqrt{(x-y)^T\Sigma^{-1}(x-y)}$$
$$dist = \sqrt{ (u-v) V^{-1} (u-v)^T } $$

å…¶ä¸­ï¼Œ(x-y)ã€(u-v)æ˜¯æ ·æœ¬ç‚¹ä¹‹é—´çš„å·®å‘é‡ï¼Œ$\Sigma^{-1}$æ˜¯åæ–¹å·®çŸ©é˜µ$\Sigma$çš„é€†çŸ©é˜µï¼Œ$T$è¡¨ç¤ºå‘é‡çš„è½¬ç½®è¿ç®—ã€‚

é©¬æ°è·ç¦»çš„è®¡ç®—è€ƒè™‘äº†ç‰¹å¾ä¹‹é—´çš„ç›¸å…³æ€§ï¼Œé€šè¿‡åæ–¹å·®çŸ©é˜µ$\Sigma$çš„é€†çŸ©é˜µæ¥å¯¹ç‰¹å¾è¿›è¡Œç¼©æ”¾å’Œæ—‹è½¬ï¼Œä½¿å¾—è·ç¦»è®¡ç®—æ›´åŠ å‡†ç¡®ã€‚å½“åæ–¹å·®çŸ©é˜µ$\Sigma$ä¸ºå•ä½çŸ©é˜µæ—¶ï¼Œé©¬æ°è·ç¦»é€€åŒ–ä¸ºæ¬§æ°è·ç¦»ã€‚

é©¬æ°è·ç¦»å¸¸ç”¨äºèšç±»åˆ†æã€å¼‚å¸¸æ£€æµ‹å’Œæ¨¡å¼è¯†åˆ«ç­‰é¢†åŸŸï¼Œç‰¹åˆ«æ˜¯åœ¨å¤„ç†å…·æœ‰ç›¸å…³ç‰¹å¾çš„æ•°æ®æ—¶ï¼Œèƒ½å¤Ÿæ›´å¥½åœ°æ•æ‰æ•°æ®ä¹‹é—´çš„çœŸå®è·ç¦»ã€‚

> è¿™é‡Œæˆ‘å¯¹é©¬æ°è·ç¦»çš„ç†è§£ä¸æ˜¯å¾ˆæ·±å…¥ï¼Œåæ–¹å·®çŸ©é˜µçš„è®¡ç®—åº”è¯¥åŸºäºæ€»ä½“æ ·æœ¬ï¼Œæ ·æœ¬é‡åº”è¯¥å¤§äºç‰¹å¾æ•°ï¼Œæ‰èƒ½ä¿è¯åæ–¹å·®çŸ©é˜µé€†çŸ©é˜µå­˜åœ¨

```python
def mahalanobis_dist(x, y, inv_cov):
    return np.sqrt(np.dot(np.dot(x-y, inv_cov), (x-y).T))


a = np.random.random(10) * 10
b = np.random.random(10) * 10
c = np.random.random(10) * 10
d = np.random.random(10) * 10
e = np.random.random(10) * 10
f = np.random.random(10) * 10
g = np.random.random(10) * 10
h = np.random.random(10) * 10
mat = np.asarray([x, y, a, b, c, d, e, f, g, h]).T
# è®¡ç®—æ€»ä½“æ ·æœ¬çš„åæ–¹å·®
cov = np.cov(mat)
# è®¡ç®—åæ–¹å·®é€†çŸ©é˜µ
VI = np.linalg.inv(np.matrix(cov))
# è®¡ç®—é©¬æ°è·ç¦»


dist = mahalanobis_dist(x, y, VI)
print(f'é©¬æ°è·ç¦» = {dist}')
dist = distance.mahalanobis(x, y, VI)
print(f'Scipyé©¬æ°è·ç¦»è·ç¦» = {dist}')
```
